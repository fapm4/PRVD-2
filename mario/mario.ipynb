## TripAdvisor Restaurants

```python
# Filtrar por Madrid
restaurants_madrid = tripadvisor[ tripadvisor['City'] == 'Madrid' ]
restaurants_madrid
```

Nos faltarían las coordenadas del restaurante, probamos a cogerlas mediante url de tripadvisor y webscrapping o con api de openstreetmaps

1. Intentar sacar dirección con web scrapping a la web de Tripadvisor

```python
# Ahora vamos a descargar una página para poder hacer web scraping
# Podemos utilizar o bien la librería requests con la funcion get (ver demo jupyter 2)
# O la librería urllib3, que se puede instalar con conda: install -c anaconda urllib3 sobre el CMD prompt de Anaconda

import urllib
from bs4 import BeautifulSoup
from time import sleep
from numpy import random

base_url_TA = 'https://www.tripadvisor.es'
#headers = {'User-Agent': 'Mozilla/6.0'}
headers = {
    "Accept": [
      "text/html,application/xhtml+xml,application/xml;q=0.9,image/avif,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7"
    ],
    "Accept-Encoding": [
      "gzip, deflate, br, zstd"
    ],
    "Accept-Language": [
      "en-US,en;q=0.9"
    ],
    "Connection": [
      "keep-alive"
    ],
    "Host": [
      "httpbin.io"
    ],
    "Sec-Ch-Ua": [
      "\"Not/A)Brand\";v=\"8\", \"Chromium\";v=\"126\", \"Google Chrome\";v=\"126\""
    ],
    "Sec-Ch-Ua-Mobile": [
      "?0"
    ],
    "Sec-Ch-Ua-Platform": [
      "\"Windows\""
    ],
    "Sec-Fetch-Dest": [
      "document"
    ],
    "Sec-Fetch-Mode": [
      "navigate"
    ],
    "Sec-Fetch-Site": [
      "none"
    ],
    "Sec-Fetch-User": [
      "?1"
    ],
    "Upgrade-Insecure-Requests": [
      "1"
    ],
    "User-Agent": [
      "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36"
    ]
  }


direcciones = [None] + len(restaurants_madrid) # Inicializar lista para guardar resultados
i = 0
for url_TA in list(restaurants_madrid['URL_TA']):
    url = base_url_TA + url_TA
    print(url)
    req = urllib.request.Request(url=url, headers=headers)
    html = urllib.request.urlopen(req).read()
    soup = BeautifulSoup(html,'html.parser')

    # Extraer la dirección
    direccion = soup.find_all('span')[60].get_text(strip=True)

    # Guardamos la dirección
    direcciones[i] = direccion
    i+=1

    break
    sleep(random.uniform(2, 4)) # meter un tiempo random entre consultas para evitar bloqueo

```

2. Intentar sacar la dirección buscando con la api de openstreetmaps

```python
# Para hacer peticiones al API importamos la librería requests
import requests
from time import sleep

api_url = 'https://nominatim.openstreetmap.org/search'


direcciones = [None] * len(restaurants_madrid) # Inicializar lista para guardar resultados
latitudes = [None] * len(restaurants_madrid) # Inicializar lista para guardar resultados
longitudes = [None] * len(restaurants_madrid) # Inicializar lista para guardar resultados

for i, row in restaurants_madrid.iterrows():
    name = row['Name']

    # Parámetros de la consulta
    params = {
        'street': name,
        'city': 'Madrid',
        'country': 'Spain',
        'type': 'restaurant',
        'addressdetails': 1,
        'limit': 1,
        'format': 'json'
    }

    # Realizar la solicitud
    response = requests.get(api_url, params=params, headers={'User-Agent': 'Mozilla/6.0'})
    
    # Verificar si la solicitud fue exitosa
    if response.status_code == 200:
        # Parsear la respuesta JSON
        data = response.json()
        if data:
            # Extraer la dirección
            direccion = data[0]['display_name']
            latitud = data[0]['lat']
            longitud = data[0]['lon']

            # Guardamos la dirección
            direcciones[i] = direccion
            latitudes[i] = latitud
            longitudes[i] = longitud
        else:
            print("No se encontraron resultados para la búsqueda.")
    else:
        print(f"Error en la solicitud: {response.status_code}")

    if i == 20:
        break
    
    sleep(1) # meter un tiempo de 1 seg para las request de la api (politica de uso de la api)


# Exportar csv con la informacion de lor restaurantes
restaurants_madrid['Address'] = direcciones
restaurants_madrid['Latitud'] = latitudes
restaurants_madrid['Longitud'] = longitudes

restaurants_madrid.to_csv('.\\TA_restaurants_processed.csv', sep=';', index=False)
```
